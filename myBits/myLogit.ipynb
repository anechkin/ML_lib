```python
import numpy as np
import matplotlib.pyplot as plt
```

## Logistic Regression

Basics of Probabilities, Odds, and Log-of-Odds (aka Logit)

```python
#given probabaility
p_success = np.arange(0.00,1.00, 0.025)
p_success = p_success[1:]
odds = p_success/(1 - p_success)
logit = np.log(odds)
print ('Probability: ', p_success)
print ('Odds: ', odds)
print ('Logit: ', logit)
```

### Plot Probabilities, Odds, Logit (Log-of-Odds)

```python
#plot probabilities and odds
plt.plot(p_success, odds)
plt.xlabel('Probabilities')
plt.ylabel('Odds')
```

```python
#plot probabilities and logit (aka log-of-odds)
plt.plot(p_success, logit)
plt.xlabel('Probabilities')
plt.ylabel('Logit')
```

```python
#plot probabilities and logit (aka log-of-odds)
plt.plot( logit, p_success)
plt.xlabel('Logit')
plt.ylabel('Probabilities')
```

```python
#plot odds and logit (aka log-of-odds)
plt.plot( odds, logit)
plt.xlabel('Odds')
plt.ylabel('Logit')
```
