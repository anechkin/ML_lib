
# The Iris Dataset

This data sets consists of 3 different types of irises'
(Setosa, Versicolour, and Virginica) petal and sepal
length, stored in a 150x4 numpy.ndarray

The rows being the samples and the columns being:
Sepal Length, Sepal Width, Petal Length and Petal Width.

The below plot uses the first two features.
See `here <https://en.wikipedia.org/wiki/Iris_flower_data_set>`_ for more
information on this dataset.


```python
print(__doc__)

# Code source: GaÃ«l Varoquaux
# Modified for documentation by Jaques Grobler
# License: BSD 3 clause

%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
# import some data to play with
iris = datasets.load_iris() #class 'sklearn.utils.Bunch', like dict

print("class: ", iris.__class__)
print("about: ", iris.keys())
print("X shape: ", iris.data.shape)
print("y shape: ", iris.target.shape)
print("Features: ", iris.feature_names)
print("Classes: ", iris.target_names)
```

### Difference between PCA with 2 features vs all 4 features

#### 2 Features

```python
#only take the first two features
X = iris.data[:, :2]
y = iris.target
```

```python
#visualize dataset, first 2 features

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(figsize=(8, 6))

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

plt.show()
```

### PCA

```python
# To getter a better understanding of interaction of the dimensions
# get the first 2 PCA dimensions
X_reduced = PCA(n_components=2).fit_transform(X)
```

```python
#visualize PCA dimensions in 1D and 2D

n = iris['data'].shape[0]

plt.figure(1, figsize=(8, 6))

# Plot 1s eigenvector in 1D
plt.scatter(X_reduced[:, 0], np.ones(n), c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.title("First PCA direction, 1D")
plt.xlabel('1st eigenvector')
plt.ylabel('n/a')
plt.yticks(())

# Plot 2 eigenvectors in 2D
plt.figure(2, figsize=(8, 6))

plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.title("First 2 PCA directions, 2D")
plt.xlabel('1st eigenvector')
plt.ylabel('2nd eigenvector')

plt.yticks(())

plt.show()
```

```python
# To getter a better understanding of interaction of the dimensions
# plot the first 2 PCA dimensions in 3D
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)

ax.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("First 2 PCA directions, 3D")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("n/a")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

Proof: No dimensionality reduction when go from 2 features to 2 eigenvectors. 

#### All 4 Features

```python
# my, expand to all 4 features
#X = iris.data
#y = iris.target
#X_reduced = PCA(n_components=3).fit_transform(X)

# To getter a better understanding of interaction of the dimensions
# get the first 3 PCA dimensions
X_reduced = PCA(n_components=3).fit_transform(iris.data)
```

```python
#visualize PCA dimensions in 1D and 2D

n = iris['data'].shape[0]

plt.figure(1, figsize=(8, 6))

# Plot 1s eigenvector in 1D
plt.scatter(X_reduced[:, 0], np.ones(n), c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.title("First PCA direction, 1D")
plt.xlabel('1st eigenvector')
plt.ylabel('n/a')
plt.yticks(())

# Plot 2 eigenvectors in 2D
plt.figure(2, figsize=(8, 6))

plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.title("First 2 PCA directions, 2D")
plt.xlabel('1st eigenvector')
plt.ylabel('2nd eigenvector')

plt.yticks(())

plt.show()
```

```python
# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimensions
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

Nice reduction from 4 features to 2 eigenvectors.
